```python
# ==================================
# MOVIE RECOMMENDATION SYSTEM
# ==================================

!pip install pandas numpy scikit-learn --quiet

import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

np.random.seed(42)

# Generate fake rating dataset
users = 100
movies = 20

ratings = np.random.randint(1, 6, size=(users, movies))

df = pd.DataFrame(ratings,
                  columns=[f"Movie_{i}" for i in range(movies)])

df.to_csv("movie_ratings.csv", index=False)

# Compute similarity between movies
similarity = cosine_similarity(df.T)

similarity_df = pd.DataFrame(similarity,
                             index=df.columns,
                             columns=df.columns)

print("Movie Similarity Matrix:")
similarity_df.head()
```
